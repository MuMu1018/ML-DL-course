```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
X = load_iris().data
X[:10]
```

```python
Y = load_iris().target
Y[:100]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def myPlot(x, labels):
    for label, c, m in zip(range(0,3),("blue","red","green"),("^","s","o")):
        plt.scatter(x[labels == label, 0], x[labels == label, 1],
                    color = c,
                    label = 'class %s' %label,
                    marker = m
                    )
    plt.show()
```

```python
myPlot(X[:,:2],Y)
```

```python
from sklearn.decomposition import PCA
# class sklearn.decomposition.PCA(n_components=None, copy=True, whiten=False,
# svd_solver=’auto’, tol=0.0, iterated_power=’auto’, random_state=None)
pca = PCA(n_components = 2)
Xpc2 = pca.fit_transform(X)
myPlot(Xpc2,Y)
```

```python
# PCA computed manually
Xcentered = X - X.mean(axis=0) # centralize
covmat = np.dot(Xcentered.T, Xcentered)
w, v = np.linalg.eig(covmat) # w eigenvalue, v eigenvector
Xpac2_manual = np.dot(Xcentered, v[:,:2]) # score Xv, projection
myPlot(Xpac2_manual,Y) # PCA may have different signs for each computation
```

```python
# IF Without centralization
covmat = np.dot(X.T, X)
w, v = np.linalg.eig(covmat) # w eigenvalue, v eigenvector
Xpac2_manual = np.dot(Xcentered, v[:,:2]) # score Xv, projection
myPlot(Xpac2_manual,Y) # PCA may have different signs for each computation

# thus, MUST DONE centralization!!
```
